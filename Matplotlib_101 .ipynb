**Basic Graph**

```python
x = [0,1,2,3,4]
y= [0,2,4,6,8]

#resize your graph
plt.figure(figsize=(5,3) , dpi=180)    # dpi is for pixels


#plt.plot(x , y , label='2x',  color='y' , linewidth=4 , linestyle='--' , marker='.' , markersize = 14 , markeredgecolor = 'blue')
plt.plot(x,y, 'y^--' , label='2x')    #easy way to draw a colorline, marker,linestyle
plt.xlabel('X axis')
plt.ylabel('Y axis' )

 # line number two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5] , x2[:5]**2 , 'r--',label='x**2')
plt.plot(x2[4:] , x2[4:]**2, 'b-',)

plt.title('My first Graph' , fontdict = {'fontname':'fantasy' , 'fontsize' : 20})
plt.legend()

plt.xticks([0,1,2,3,4,])
plt.yticks([1.5 , 3 , 4.5])

plt.savefig('plot.png')

plt.show()

```

### **Histograms**

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40 , 50 , 60 , 70 , 80 , 90 , 100]

plt.hist(fifa.Overall , bins = bins , color='#abcdef')

plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')

plt.title('Distribution of players skills in fifa 2022')
plt.show()
```

## **Pie Chart**

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
left
```

```python
right =fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
right
```

```python
labels=['left' , 'right']
colors = ['#abcdef','#abdefd']

plt.pie([left , right] , labels= labels , colors=colors, autopct='%.2f %%')    #the last one is to fill the percentage automaticaly

plt.title('Foot Preferrence of Fifa Players')
plt.show()
```

```python
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else  x for x in fifa.Weight]
fifa.Weight
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125)&(fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150)&(fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175)&(fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]

plt.style.use('ggplot')

weights = [light , light_medium , medium , medium_heavy , heavy]
labels = ['light','light_medium','medium','medium_heavy','heavy']
explode=(.4 , .1 , 0 , 0 , .4)

plt.pie(weights , labels=labels , autopct= '%.2f %%' , pctdistance=0.9 , explode = explode)
plt.title('Weight Distribution of Fifa Players')
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

labels=['Barcelona', 'Madrid']
boxes = plt.boxplot([barcelona , madrid] , labels = labels , patch_artist = True , medianprops={'linewidth':2})

for box in boxes['boxes']:      # boxes that is inside is a parameter
    # set edge coor and width
  box.set(color='#4286f4' , linewidth=2)
    #change fill color
  box.set(facecolor='#e0e0e0')

plt.style.use('default')
plt.title('Professional foot teams comparaison')
plt.ylabel('Fifa Overall Ranking ')

plt.figure(figsize=(5,8))
plt.show()
```
